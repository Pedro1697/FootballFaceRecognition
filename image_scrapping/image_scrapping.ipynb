```python
import os
from icrawler.builtin import GoogleImageCrawler

# Directory
drive_dir = "/content/drive/My Drive/Portfolio/soccer_player_classifier/model/img_dataset"

# Player's list
players_list = ["Ronaldo", "Messi", "Neymar", "Mbappe", "Haaland"]

# Dowload the images
for player in players_list:
    print(f"Dowloanding the images of {player}...")

    player_dir = os.path.join(drive_dir, player)
    os.makedirs(player_dir, exist_ok=True)

    crawler = BingImageCrawler(
        storage={"root_dir": player_dir},
        downloader_threads=4,
        parser_threads=2
    )


    search_query = f"{player} football player"

    crawler.crawl(
        keyword=search_query,
        max_num=700,
        min_size=(200, 200),  # Min size of images
        max_size=None
    )

print("Downloading complete. The images are in the directory.")
```
